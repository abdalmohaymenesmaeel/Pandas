<h2>Loading CSV File</h2>

<p>
    <ul>
        <li><b>sep:</b>By default, pd.read_csv() assumes that the delimiter is a comma (,), but you can specify a different delimiter using the sep parameter</li>
                <li><b>skiprows:</b>You can skip a specific number of initial rows using the skiprows parameter. This is useful when your CSV file has header information or comments at the beginning</li>
        <li><b>header:</b>By default, pandas assumes the first row of the CSV file contains column names. If your file doesn't have a header, you can specify it as None and pandas will generate default column names</li>
        <li><b>name:</b>You can explicitly specify column names using the names parameter</li>
        <li><b>index_col:</b>You can specify which column to use as the DataFrame index using the index_col parameter. For example, if the first column contains unique identifiers</li>
                <li><b>na_values:</b>You can specify how missing values are represented in your CSV file using the na_values parameter</li>
        <li><b>encoding:</b>If your CSV file uses a different encoding, you can specify the encoding explicitly using the encoding parameter</li>
    </ul>
</p>

```python
df = pd.read_csv('students.csv', sep=',')
df = pd.read_csv('students.csv', skiprows=2)
df = pd.read_csv('students.csv', header=None)
# Define custom column names
column_names = ['unumber', 'firstname', 'City']
df = pd.read_csv('students.csv', names=column_names)
# Use the 'ID' column as the index
df = pd.read_csv('students.csv', index_col='unumber')
# Treat 'N/A' and 'Missing' as NaN values
df = pd.read_csv('students.csv', na_values=['N/A', 'Missing'],index_col='unumber')
#df = pd.read_csv('students.csv', encoding='utf-8') 
df
```

<h1>Joining</h1>

```python
# Create two sample DataFrames
data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'ID': [2, 3, 4], 'Age': [25, 30, 35]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Merge the DataFrames using an inner join on 'ID'
result = pd.merge(df1, df2, on='ID', how='inner')

# Display the merged DataFrame
result
```

```python
# Merge using a left join
result = pd.merge(df1, df2, on='ID', how='left')
#select * from students s1 left join students s2 on s1.Id=s2.Id
# Display the merged DataFrame
result
```

```python
# Merge using a right join
result = pd.merge(df1, df2, on='ID', how='right')

# Display the merged DataFrame
result
```

```python
# Merge using an outer full join
result = pd.merge(df1, df2, on='ID', how='outer')

# Display the merged DataFrame
result
```

<h1>Ploting</h1>

```python

# Create a sample DataFrame
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Sales': [100, 150, 200, 180, 250]}
df = pd.DataFrame(data)

# Create a line plot
df.plot(x='Year', y='Sales', kind='line')
plt.title('Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Create a bar plot
df.plot(x='Year', y='Sales', kind='bar')
plt.title('Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Create a histogram
df['Sales'].plot(kind='hist', bins=5)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a scatter plot
df.plot(x='Year', y='Sales', kind='scatter')
plt.title('Scatter Plot: Sales vs. Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Create a box plot
df['Sales'].plot(kind='box')
plt.title('Box Plot: Sales')
plt.ylabel('Sales')
plt.show()
```
